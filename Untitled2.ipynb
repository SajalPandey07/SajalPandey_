```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import datetime as dt

# 1. DATA CLEANING & FEATURE ENGINEERING (RFM)
# Load dataset
df = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1')

# Remove rows without CustomerID and define TotalAmount
df = df.dropna(subset=['CustomerID'])
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

# Calculate Recency, Frequency, and Monetary (RFM)
snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalAmount': 'sum'
})
rfm.rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalAmount': 'Monetary'}, inplace=True)

# Filter out non-positive Monetary values (returns exceeding purchases)
rfm = rfm[rfm['Monetary'] > 0]

# 2. K-MEANS CLUSTERING
# Log transform to handle skewness and Scale the data
rfm_log = np.log1p(rfm)
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)

# Fit K-Means with K=3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# 3. VISUALIZATION USING PCA
pca = PCA(n_components=2)
rfm_pca = pca.fit_transform(rfm_scaled)
pca_df = pd.DataFrame(rfm_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = rfm['Cluster'].values

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis')
plt.title('Customer Segments Visualization (PCA)')
plt.savefig('pca_clusters.png')

# 4. LABEL CLUSTERS
# Identify labels based on mean Monetary value
cluster_summary = rfm.groupby('Cluster')['Monetary'].mean().sort_values()
labels = {cluster_summary.index[0]: 'Inactive',
          cluster_summary.index[1]: 'Occasional',
          cluster_summary.index[2]: 'Loyal/High-Value'}
rfm['CustomerLabel'] = rfm['Cluster'].map(labels)

# 5. SUPERVISED CLASSIFICATION (SVM)
# Define High-Value as top 25% of Monetary value
rfm['IsHighValue'] = (rfm['Monetary'] >= rfm['Monetary'].quantile(0.75)).astype(int)
X = rfm_scaled
y = rfm['IsHighValue']

# Model 1: Without Cluster Labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm_no_cluster = SVC(kernel='linear').fit(X_train, y_train)
acc_no = accuracy_score(y_test, svm_no_cluster.predict(X_test))

# Model 2: With Cluster Labels
X_with_c = np.column_stack((rfm_scaled, pd.get_dummies(rfm['Cluster']).values))
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_with_c, y, test_size=0.3, random_state=42)
svm_with_cluster = SVC(kernel='linear').fit(X_train_c, y_train_c)
acc_with = accuracy_score(y_test_c, svm_with_cluster.predict(X_test_c))

# Output Results
print(f"Cluster Interpretation:\n{rfm.groupby('CustomerLabel')[['Recency', 'Frequency', 'Monetary']].mean()}")
print(f"\nAccuracy Without Cluster Features: {acc_no:.4f}")
print(f"Accuracy With Cluster Features: {acc_with:.4f}")
```
